# Example of fitting a one-dimensional MDN  with one output
-------

We use the pydra module to construct a one-dimensional MDN with a single input and test it on unimodal generated data

The data $y$ is drawn from a normal distribution for input $x$, where $y \sim N(\mu(x),\sigma^2)$. The functional form of the mean is an inverse quadratic i.e. $\mu(x) = -ax^2$.

```python
model = pydra.load_mdn_model(cluster_size=10,output_size=2,layers=3,input_size=1,
                       dense_layer_size=64,print_summary=True)
```

## Set-up data

```python
x = np.random.uniform(0,1,1000)
```

```python
_,y1 = gd.final_size(x=x)
_,y2 = gd.u_shape(x=x)
```

```python
plt.subplot(211)
plt.plot(x,y1,'o');
plt.subplot(212)
plt.plot(x,y2,'o');
```

```python
x = x.reshape(x.size,1)
```

```python
history = model.fit(x, [y1, y2], epochs=500, batch_size=500,
          verbose=1)
```

## Plot Loss for each output

```python
print(history.history.keys())
```

```python
for output in ['output_{}_loss'.format(i) for i in range(2)]:
    plt.plot(history.history[output],label=output);
plt.legend();
```

## Test output results

```python
x_test = np.float32(np.arange(0,1.0,0.0001))
x_test = x_test.reshape(x_test.size,1)
```

```python
output = model.predict(x_test)
```

```python
for i,y in enumerate([y1,y2]):
    y_test = pydra.generate_mdn_sample_from_ouput(output[i], x_test.size,num_components=10)
    plt.figure(figsize=(8, 8))
    
    plt.plot(x_test,y_test,'bo',alpha=0.1,label='mdn output')
    plt.plot(x,y,'ro',alpha=0.3,label='data')
    plt.legend();plt.xlabel('x');plt.ylabel('y');
```
