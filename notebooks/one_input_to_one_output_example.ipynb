# Example of fitting a one-dimensional MDN  with one output
-------

We use the pydra module to construct a one-dimensional MDN with a single input and test it on unimodal generated data

The data $y$ is drawn from a normal distribution for input $x$, where $y \sim N(\mu(x),\sigma^2)$. The functional form of the mean is an inverse quadratic i.e. $\mu(x) = -ax^2$.

## Instantiate mdn model

```python
model = pydra.load_mdn_model()
```

## Generate fake data

```python
x = np.float32(np.random.uniform(0,1,(1, 1000))).T
_,y = gd.u_shape(x=x)
```

```python
plt.figure(figsize=(8,8));
plt.plot(x,y,'o',alpha=0.5); plt.xlabel('x'); plt.ylabel('y');
```

## Fit model

```python
history = model.fit(x, y, batch_size=200, epochs=500, verbose=1)
```

```python
print(history.history.keys())
```

```python
plt.figure(figsize=(12,8));
plt.plot(history.history['loss']); plt.xlabel('Epochs'); plt.ylabel('loss');
```

## Create test data and compare to MDN model

```python
x_test = np.float32(np.arange(0.,1.0,0.001))
x_test = x_test.reshape(x_test.size,1)
output = model.predict(x_test)
pu.plot_samples(output,y,x,x_test)
```

```python
pu.plot_mean_and_var(output,x_test)
```

# Fit to bimodal data
--------

The idea behind this data is to simulate a process that would look similar to the final size of an epidemic as the $R_0$ increases.

```python
x = np.random.uniform(0,1,1000);
_,y = gd.final_size(x=x)
```

```python
plt.figure(figsize=(8,8));
plt.plot(x,y,'o');
```

```python
x = x.reshape(x.size,1)
```

```python
model = pydra.load_mdn_model()
```

```python
history = model.fit(x, y, batch_size=200, epochs=300, verbose=1)
```

```python
plt.figure(figsize=(12,8));
plt.plot(history.history['loss']); plt.xlabel('Epochs'); plt.ylabel('loss');
```

```python
x_test = np.float32(np.arange(0.,1.0,0.001))
x_test = x_test.reshape(x_test.size,1)
output = model.predict(x_test)
pu.plot_samples(output,y,x,x_test)
```

```python
pu.plot_mean_and_var(output,x_test)
```
